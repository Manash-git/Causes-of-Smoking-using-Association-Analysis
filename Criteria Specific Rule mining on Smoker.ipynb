## Rule mining on " Reasons or influences of starting smoking"

### Final data processing

```python
# smoker['reasons or influences of start smoking?'].unique()
smoker['reasons or influences of start smoking?'].unique().shape
```

```python
# final casting data to string

smoker['reasons or influences of start smoking?'] = smoker['reasons or influences of start smoking?'].astype(str)
smoker['index']= smoker['index'].astype(str)
```

```python
model_data = list(smoker['reasons or influences of start smoking?'].apply(lambda x:x.split(',')))
model_data
```

### Applying Apriori

```python
# importing libraries

from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_data = te.fit(model_data).transform(model_data)
te_data
```

```python
df = pd.DataFrame(te_data,columns=te.columns_)
df.head()
```

```python
frq_items = apriori(df, min_support = 0.1, use_colnames = True)

frq_items
```

```python
rules = association_rules(frq_items, metric ="confidence", min_threshold = .6) 
# rules.shape
rules
```

### Againg importing libraries for separte mining

```python
# importing libraries

from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

## Rule mining on smokers influencing factor for continuing smoking

```python
# Casting

smoker['continuing cigarette smoking?'] = smoker['continuing cigarette smoking?'].astype(str)
smoker['index']= smoker['index'].astype(str)
```

```python
# rule mining

model_data = list(smoker['continuing cigarette smoking?'].apply(lambda x:x.split(',')))
# model_data
te = TransactionEncoder()
te_data = te.fit(model_data).transform(model_data)
# te_data
df = pd.DataFrame(te_data,columns=te.columns_)
frq_items = apriori(df, min_support = 0.1, use_colnames = True)

frq_items
```

```python
rules = association_rules(frq_items, metric ="confidence", min_threshold = .6) 
# rules.shape
rules = rules.sort_values( ['confidence','support'], ascending= [False, False] )
rules.head()
```

## Rule mining on factors that helps smoker to overcome

```python
# Casting

smoker['Smoking helps?'] = smoker['Smoking helps?'].astype(str)
smoker['index']= smoker['index'].astype(str)
# rule mining

model_data = list(smoker['Smoking helps?'].apply(lambda x:x.split(',')))
# model_data
te = TransactionEncoder()
te_data = te.fit(model_data).transform(model_data)
# te_data
df = pd.DataFrame(te_data,columns=te.columns_)
frq_items = apriori(df, min_support = 0.1, use_colnames = True)

frq_items
```

```python
rules = association_rules(frq_items, metric ="confidence", min_threshold = .6) 
# rules.shape
rules = rules.sort_values( ['confidence','support'], ascending= [False, False] )
rules.head(20)
```
