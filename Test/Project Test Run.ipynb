```python
model_data = list(data['reasons or influences of start smoking?'].apply(lambda x:x.split(',')))
model_data
```

```python
# data.head()
```

```python
f_data= ( data.groupby(['index','reasons or influences of start smoking?']))
# f_data.head()
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
```

```python
te_data = te.fit(model_data).transform(model_data)
```

```python
te_data
```

```python
te.columns_
```

```python
df = pd.DataFrame(te_data,columns=te.columns_)
df.head()
```

```python
frq_items = apriori(df, min_support = 0.1, use_colnames = True)

frq_items
```

```python
rules = association_rules(frq_items, metric ="confidence", min_threshold = .6) 
# rules.shape
rules
```
